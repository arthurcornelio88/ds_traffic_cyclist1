## ‚û§ 3. Nettoyage de donn√©es manquantes, duplicates

```python
# V√©rification des doublons
df.duplicated().sum()
```

```python
# Pourcentage de valeurs manquantes par colonne
missing_percent = df.isnull().mean() * 100
missing_percent[missing_percent > 0].sort_values()
```

```python
# Traitement des dates
df['Date et heure de comptage'] = pd.to_datetime(df['Date et heure de comptage'], errors='coerce')
df["Date d'installation du site de comptage"] = pd.to_datetime(df["Date d'installation du site de comptage"], errors='coerce')
# Nettoyage des coordonn√©es et extraction
df[['lat', 'lon']] = df['Coordonn√©es g√©ographiques'].str.extract(r'([-\d.]+)[,\s]+([-\d.]+)').astype(float)
```

## ‚û§ 4. Analyse des distributions

```python
# Histogrammes des variables num√©riques
df.hist(figsize=(14, 10), bins=30)
plt.tight_layout()
plt.show()
```

```python
# Distribution des cyclistes (modifie le nom de la colonne cible si besoin)
sns.histplot(df['nb_cyclistes'], kde=True)
plt.title("Distribution du trafic de cyclistes")
plt.show()
```

## ‚û§ 5. D√©tection des outliers

```python
# Bo√Ætes √† moustaches (Boxplots)
num_cols = df.select_dtypes(include=np.number).columns

for col in num_cols:
    plt.figure(figsize=(8, 3))
    sns.boxplot(x=df[col])
    plt.title(f"Outliers pour {col}")
    plt.show()
```

## ‚û§ 6. Traitement de donn√©es

```python
# Exemple de transformation de dates (si colonne "date")
if "date" in df.columns:
    df["date"] = pd.to_datetime(df["date"])
    df["jour_semaine"] = df["date"].dt.day_name()
    df["mois"] = df["date"].dt.month
```

## ‚û§ 7. Corr√©lations

```python
# Matrice de corr√©lation
plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title("Matrice de corr√©lation")
plt.show()
```
